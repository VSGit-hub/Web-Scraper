## Importing packages

```python
from requests import get
from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep, time
import pandas as pd
```

## Source code

```python
# Creating lists for storing Data
post_date = []
quest_number = []
category_code = []
bid_name = []
bid_closing_date = []
city = []
country = []
state = []
owner = []
solicitor = []
posting_type =[]

colunms = [post_date, quest_number, category_code, bid_name, bid_closing_date, city, country, state, owner,solicitor, posting_type]

# Base URL
base_url = 'https://qcpi.questcdn.com/cdn/posting/?group=1950787&provider=1950787'

def get_info(element_identifier,list_name) :
    if element_identifier is not None :
        list_name.append(element_identifier)
    else :
        list_name.append("No info") 
        
def get_text(index,td_number) :
        return index.find_all('td')[td_number].text        
        
# Iterating through pages
pages = [str(i) for i in range(1)]
for page in pages:
    browser=webdriver.Chrome()
    browser.maximize_window()
    browser.get(base_url)      
    
    total_height = int(browser.execute_script("return document.body.scrollHeight"))
    for i in range(1, total_height, 500):
        browser.execute_script("window.scrollTo(0, {});".format(i))
        sleep(4)     
    
    
    # Parse the content of the request with BeautifulSoup
    html = browser.page_source
    html_soup = BeautifulSoup(html,'html.parser')

    # Select all tabel data
    table = html_soup.find(id = 'table_id')
    table_rows = table.tbody.find_all('tr')

    row_number = 0
    # Extract data for individula row 
    for data in table_rows:
        
        for i in range(0,11):
            get_info(get_text(data,i),colunms[i])
        
        row_number += 1
        
        if row_number == 10:
            break
        clear_output(wait = True)
    browser.close()
```

## Converting to data frame

```python
pd.options.display.float_format = '{:.2f}'.format 
df = pd.DataFrame({
   "Post Date": post_date,
    "Quest Number":quest_number, 
    "Category Code":category_code, 
    "Bid_name":bid_name, 
    "Bid Closing Date":bid_closing_date, 
    "City":city, 
    "Country":country, 
    "State":state, 
    "Owner":owner,
    "Solicitor":solicitor, 
    "Posting Type":posting_type

})       
print(df.info())
df 
```
